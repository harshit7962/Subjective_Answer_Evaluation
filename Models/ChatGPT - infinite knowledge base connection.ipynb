```python
def getModelAnswer(question):
    openai.api_key = "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"
    context = "DBMS"
    response = openai.Completion.create(
      model="text-davinci-003",
      prompt=question + " in " + context,
      temperature=0.5,
      max_tokens=1000,
      top_p=1.0,
      frequency_penalty=0.8,
      presence_penalty=0.0
    )

    response = response.choices[0].text.replace("\n"," ")
    response = response[2:]
    return response
```
